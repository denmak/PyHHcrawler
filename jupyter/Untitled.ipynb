```python
import  csv
from urllib.parse import urlparse
from urllib.parse import parse_qs
from urllib.parse import urlencode
import re
import pandas

def read1(path):
    with open(path,encoding='utf-8') as csvfile:
        readcsv = csv.reader(csvfile,delimiter = ',')
        i = 1
        df =pandas.DataFrame()
        for row in readcsv:
            if i>1:
                print(row[2])
                url = row[5]
                o = urlparse(url)
                query = parse_qs(o.query)
                regex=  re.compile('\d+')
                
                print(i,' ',regex.findall(row[2].replace(' ',''))[0],query.get('text')[0])
            i=i+1

def format_row(row,col_num): 
    regex=  re.compile('\d+')
    return regex.findall(row[col_num].replace(' ',''))[0]

def format_row2(row):
    o = urlparse(row[5])
    query = parse_qs(o.query)
    return query.get('text')[0]

def from_count(row,col_num,splt,nm=0,tp=0):
    if nm == 0:
        s = row[col_num].split(splt)        
        return len(s)
    else:
        s = row[col_num].replace('руб','').split(splt)          
        return s[nm].split('.')[tp].replace(' ','')
    
def from_city(row,col_num,splt,nm=0,xc='\D+'):
    if nm == 0:
        s = row[col_num].split(splt)          
        return len(s)
    else:
        s = row[col_num]
        regex =  re.compile(xc)
        return regex.findall(s)[nm]
              
def read2(path):
    df = pandas.read_csv(filepath_or_buffer = path,header=0)
    df['total']       = df.apply(lambda row: format_row(row,2),axis=1)
    df['with_salary'] = df.apply(lambda row: format_row(row,4),axis=1)
    df['lang']        = df.apply(lambda row: format_row2(row),axis=1)
    for i in range(1,6):
        col_grade           ='grade'+str(i)
        col_grade_count     ='with_grade'+str(i)+ '_count'
        df[col_grade]       = df.apply(lambda row: from_count(row,4,'от',i),axis=1)
        df[col_grade_count] =df.apply(lambda row: from_count(row,4,'от',i,1),axis=1)
    for i in range(1,21):
        col_city           ='top20_city'+str(i)
        col_city_count     ='top20_city'+str(i)+ '_count'
        df[col_city]       = df.apply(lambda row: from_city(row,1,' ',i),axis=1)
        df[col_city_count] = df.apply(lambda row: from_city(row,1,' ',i,'\d+\s'),axis=1)  
    df.drop(df.columns[[0,1,2,3,4,5]], axis=1, inplace=True) 
    df.to_csv(path.replace('hh3','featured_hh3'),encoding='utf-8')
    print(df)
    
            
read2('C:\\Users\\Acer\\Downloads\\hh39thMar23_30.csv')            
```
